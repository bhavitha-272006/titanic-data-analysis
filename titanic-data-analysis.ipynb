```python
## 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
## 2. Load Data
data=pd.read_csv("/content/Titanic-Dataset.csv")
df=pd.DataFrame(data)
print(df)
print(df.head())
print(df.tail())
print(df.shape)
print(df.columns)
print(df.info())
print(df.isnull().sum())
print(df.describe())
```

```python

## 3. Data Cleaning
df['Age']=df['Age'].fillna(df['Age'].median())
df=df.drop('Cabin',axis=1)
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
print(df.isnull().sum())
```

```python
## 4. Exploratory Data Analysis
sns.boxplot(data=df,x='Pclass',y='Fare')
```

```python
sns.heatmap(df[['Survived','Age','Fare','Pclass','SibSp','Parch']].corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title("correlation matrix")
plt.show()
```

```python
## Conclusions:
#1. Only 38% of passengers survived
#2. Females had much higher survival rate than males
#3. 1st class passengers survived more than 3rd class
#4. Higher fare = better survival chance
#5. Age was not a strong factor in survival
```
